# PCA

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.preprocessing import scale 
from sklearn import decomposition
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report
```

```python
df_pca = df.copy()
```

Testing 

```python
#Create testing and training sets 
train = df_pca.sample(frac=8/10, random_state = 0)
test = df_pca.drop(train.index)

train_x = train.drop(['transportation_issues'], axis=1)
train_y = train['transportation_issues']

test_x = test.drop(['transportation_issues'], axis=1)
test_y = test['transportation_issues']
```

```python
scaler = StandardScaler()

scaler.fit(train_x)
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)
```

```python
#PCA
pca = PCA(n_components = 20, whiten='True')
pca.fit_transform(train_x)
```

```python
# Train Model to X and Y
logisticRegr = LogisticRegression(solver = 'liblinear', random_state=0) #liblinear or lbfgs for solver
logisticRegr.fit(train_x, train_y)
```

```python
#Predicting test_x
y_pred = logisticRegr.predict(test_x) 
```

```python
#Review results
logisticRegr.score(test_x, test_y)
```

```python
accuracy_score(test_y, logisticRegr.predict(test_x)) 
```

```python
confusion_matrix(test_y, y_pred) 
```

```python
print(classification_report(test_y,y_pred))
```

```python
#Making data frame with possibility of 1/0
df_y_mod=logisticRegr.predict_proba(test_x)
df_y_prop=pd.DataFrame(df_y_mod, columns=['prop_no_tran_issues_','prop_transportation_issues'])
print(df_y_prop.shape)
```

```python
df_y_pred=pd.DataFrame(y_pred, columns=['transportation_issues'])#prediction data set 
print(df_y_pred.shape)
```

```python
#Creating a column for the correct/original transportation_issues
y_test = test[['transportation_issues']]
y_test = y_test.reset_index(drop=True)
y_test.columns = ['original_transportation_issues']
```

```python
#Merging prediction, probability, and original to compare 
df_y_merge = pd.concat([df_y_pred, df_y_prop, y_test], axis=1, sort=False)
df_y_merge.head(50)
```

```python
# Filter by transportation_issues = 1 and sorting by highest propability 
df_y_merge = df_y_merge[df_y_merge['transportation_issues'] == 1].sort_values(by=['prop_transportation_issues'], ascending = False)
df_y_merge.head(10)
```
